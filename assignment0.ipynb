```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 400)
y = np.sin(x)
plt.figure(figsize=(10, 6))
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.plot(x, y)
plt.show()
```
